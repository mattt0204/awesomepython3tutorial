- 객체의 메서드를 선언시점이 아닌 사용시점에 확장하는 것을 몽키 패치라고 합니다.

import할 때 몽키패칭 주의하세요.
- 동적으로 추가 되서 늘어난다(변수, 메서드 다 .. ) 이게 바로 몽키패칭
- slot을 사용하면 slot에 정의되어 있는 애들만 사용할 수 있다. (몽키패칭 방지)

# print를 사용하는 경우와 안 하는 경우
- \_\_repr__ : raw format, escape 자동 생성 (개발자들이 선호하는 포맷)
- \_\_str__ : formatting

```python
class B:
    def __repr__(self):
        return '파이썬'
    def __str__(self):
        return '공부중'
```

```python
b = B()
```

```python
b
```

```python
a = r'asdf\n'
a
```

```python
print(a)
```

```python
a = 'asdf\n'
a
```

```python
print(a)
```

```python
print(b)
```

# \_\_getattr__
- a 객체 내에 없는 변수인 var를 호출하면 자동으로   \_\_getattr__()을 실행한다

```python
class X:
    def __getattr__(self,attr):
        return '파이썬프로젝트'
```

```python
a= X()
```

```python
dir(a)
```

```python
a.a
```

```python
a.v
```

```python
class X:
    def __getattr__(self,attr):
        if attr == 'a':
            return '파이썬'
        return '공부중'
```

```python
a = X()
```

```python
a.a
```

```python
a.b
```

```python
a = 1.0
a.hex()
```

```python
getattr(a, 'hex')()
```

```python
a.__getattribute__('hex')() # .으로 탐색했을 때 사용할 수 있는것
```

```python
try:
    a = 1/0
except:
    a = 2/0
```

# 덕타이핑 예)
- \_\_getitem__이 있으면 iterable로 사용할 수 있습니다.
- iterable을 상속하지 않고도 iterable로 사용할 수 있습니다.

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
dir(tips)
```

```python
'__getitem__' in dir(tips)
```

```python
for i in tips.iterrows():
    print(i)
```
